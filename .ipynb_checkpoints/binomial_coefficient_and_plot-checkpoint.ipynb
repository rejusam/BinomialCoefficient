```python
import numpy as np
from decimal import *
import scipy.special
from scipy.special import factorial
import pandas as pd
from math import e
from math import pi
import matplotlib.pyplot as plt
import seaborn as sns


getcontext().prec = 1000

n = 100
c = []
for i in range(0,n+1):
    a = sum(scipy.special.comb(Decimal(i), (r), exact=True)*pow(-1, r)/Decimal(factorial(r, exact=True)) for r in range(i+1))
    c.append(a)



df_dist = {'i':range(0, n+1), 'sum': c}

df = pd.DataFrame(df_dist)
df = df.set_index('i')
df.to_csv('tiju_summation_100.csv', index=True)
```

```python
df_1 = pd.read_csv('tiju_summation_5000.csv') # , index_col='i'
df_1['sum absolute square'] = df_1['sum'].abs()**2
df_1['second function'] = (1/(pow(e, 1/12)*2*pi*df_1['i']))*(pow((1 -e/df_1['i']),2*df_1['i']))
df_1 = df_1.set_index('i', drop=True)
df_1 = df_1.drop([0], axis=0)
#display(df_1)


plt.rcParams['figure.figsize'] = [20, 12] #[W,H]
plt.rcParams['figure.dpi'] = 140
plt.rcParams['text.usetex'] = True
plt.style.use('ggplot')
ax = df_1.plot(
    y=["sum absolute square", "second function"],
    style="o--",
    title=r"$\displaystyle\sum_{r=0}^i \frac{i!}{r!(i-r)!} \frac{(-1)^{r}}{r!}$ vs. $i$ plot",
    xlabel=r"$i$",
    ylabel=r"estimation",
    figsize=(20, 12),
    #xlim=[start_time, end_time],
    ylim=[0, 0.5],
)

#ax.set_yscale('log')
#ax.set_xticks(range(len(df_1.index)));
#ax.set_xticklabels(df_1.index);
#plt.xticks(rotation=90);

plt.savefig('summation.pdf', 
            metadata = {'Creator': 'Reju Sam John', 'Author': 'Reju', 'Title': 'binomial-coefficient'},
           bbox_inches = 'tight', pad_inches = 0.2)
plt.show()
```

```python

plt.rcParams['figure.figsize'] = [20, 12] #[W,H]
plt.rcParams['figure.dpi'] = 140
plt.rcParams['text.usetex'] = True
plt.style.use('ggplot')

# plot line chart on axis #1
ax1 = sns.lineplot(
    x=df_1.index, 
    y='sum absolute square', 
    data=df_1, 
    sort=False, 
    color='blue'
)
ax1.set_ylabel(r"$\displaystyle\sum_{r=0}^i \frac{i!}{r!(i-r)!} \frac{(-1)^{r}}{r!}$")
#ax1.set_ylim(0, 0.5)
#ax1.set_yscale('log')
ax1.legend([r"$\displaystyle\sum_{r=0}^i \frac{i!}{r!(i-r)!} \frac{(-1)^{r}}{r!}$"], loc="upper center")
ax1.yaxis.label.set_color('blue')
ax1.yaxis.label.set_fontsize(14)
ax1.tick_params(axis='y', colors='blue', labelsize=14)

# set up the 2nd axis
ax2 = ax1.twinx() 

# plot bar chart on axis #2
sns.lineplot(
    x=df_1.index, 
    y='second function', 
    data=df_1, 
    sort=False, 
    color='red', 
    ax = ax2       # Pre-existing axes for the plot
)
ax2.grid(False) # turn off grid #2
ax2.set_ylabel(r"$\displaystyle\frac{1}{e^{1/12}2\pi i} \left(1 - \frac{e}{i}\right)^{2i}$ (in log scale)")
ax2.set_ylim(1e-8, 5e-01)
ax2.set_yscale('log')
ax2.legend([r"$\displaystyle\frac{1}{e^{1/12}2\pi i} \left(1 - \frac{e}{i}\right)^{2i}$"], loc="best")
ax2.yaxis.label.set_color('red')
ax2.yaxis.label.set_fontsize(14)
ax2.tick_params(axis='y', colors='red', labelsize=14)



plt.title(r"$\displaystyle\sum_{r=0}^i \frac{i!}{r!(i-r)!} \frac{(-1)^{r}}{r!}$ and $\displaystyle\frac{1}{e^{1/12}2\pi i} \left(1 - \frac{e}{i}\right)^{2i}$ vs. $i$ plot")

plt.savefig('summation_diff_scale_i_5000.pdf', 
            metadata = {'Creator': 'Reju Sam John', 'Author': 'Reju', 'Title': 'binomial-coefficient'},
           bbox_inches = 'tight', pad_inches = 0.2)

plt.show()
```
